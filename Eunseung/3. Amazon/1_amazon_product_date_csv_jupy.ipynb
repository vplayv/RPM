```python
import pandas as pd
import numpy as np
import dateutil
import calendar

# package to extract data from various Internet sources into a DataFrame
# make sure you have it installed
from pandas_datareader import data, wb

# package for dates
import datetime as dt
```

```python
A_vendor_weigh = pd.read_csv('Amazon_0822_0333_segonii.csv')
```

```python
#데이터 타임으로 변경
A_vendor_weigh['Date'] = pd.to_datetime(A_vendor_weigh.loc[:,'Date'])
```

```python
#두 컬럼만 불러오기
A_vendor_weigh = A_vendor_weigh[['Date', 'Product_name']]
A_vendor_weigh.head()
```

```python
# 내림차순으로 정렬
A_vendor_weigh = A_vendor_weigh.sort_values(['Date'], ascending=[False])
```

```python
# 데이트 뉴, 앞으로 설정
A_vendor_weigh = A_vendor_weigh.set_index('Date')
```

```python
#주차별로 멀티 인덱스 만들기
A_vendor_weigh1 = A_vendor_weigh.groupby(['Product_name','Date']).resample('w').count().unstack()
```

```python
#결측값 = 0 처리
A_vendor_weigh1 = A_vendor_weigh1.fillna(0)
```

```python
#데이트 뉴 행으로 합치기 
A_vendor_weigh2 = A_vendor_weigh1.sum(axis=1, level = 'Date')
```

```python
#주+벤더별 정렬하기
A_vendor_weigh3 = A_vendor_weigh2.sum(axis=0, level = 'Product_name')
```

```python
# 칼럼명 변경
A_vendor_weigh3.columns = ["Week30", "Week31", "Week32", "Week33"]
```

```python
#칼럼명으로 오름차순 정렬

A_vendor_weigh3 = A_vendor_weigh3.sort_index(axis =1 , ascending = True)
```

```python
####################################################
#####주차별 가중치합 (기존의 Weight_sum과 달리, 2~4주차치만 뽑은 데이터!)

A_vendor_weigh3['Weight_sum'] = A_vendor_weigh3['Week33'] + A_vendor_weigh3['Week32'] + A_vendor_weigh3['Week31'] + A_vendor_weigh3['Week30']
```

```python
#내림차순으로 정렬
A_vendor_weigh3 = A_vendor_weigh3.sort_values(["Weight_sum"], ascending=[False])
A_vendor_weigh3
```

```python
# csv파일로 저장하기
data = pd.DataFrame(A_vendor_weigh3)
data.to_csv('###Amazon_product_date_sum.csv')
```
