```python
#띄어쓰기 별로 나누고 불러오기
samsung_ms['Product_name'] = samsung_ms['Product_name'].str.split(' ').str[:3]
samsung_ms
```

```python
#삼성제품만 불러오기
samsung_ms = samsung_ms[samsung_ms['Product_name'].str[0] == 'Samsung']
samsung_ms
```

```python
#제품 이름만 불러오기 
samsung_ms['Product_name'] = samsung_ms['Product_name'].str[2]
samsung_ms
```

```python
#제품 시리즈만 불러오기
samsung_ms['Product_name'] = samsung_ms['Product_name'].str[0]
samsung_ms
```

```python
#진짜 제품라인으로 이름 변경
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('A','A series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('J','J series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('M','M series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('O','ON series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('P','ON series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('S','S series')
samsung_ms['Product_name'] = samsung_ms['Product_name'].replace('NOTE','S series')


samsung_ms['Product_name'].head()
```

```python
# 제품라인별로 그룹바이
samsung_ms_prod = samsung_ms.groupby(['Product_name']).sum()
samsung_ms_prod
```

```python
samsung_ms_prod = samsung_ms_prod.drop(['E', 'G', 'C' , 'N'])
samsung_ms_prod
```

```python
samsung_ms_prod.columns = ["Week34","Week33", "Week32", "Week31" ,"Week30", "Weight_sum"]

#칼럼명으로 오름차순 정렬
samsung_ms_prod = samsung_ms_prod.sort_index(axis =1 , ascending = True)
samsung_ms_prod
```

```python
# ####################################################
# #####주차별 가중치합

# samsung_ms_prod['Weight_sum'] = samsung_ms_prod['Week30'] + samsung_ms_prod['Week31'] + samsung_ms_prod['Week32'] + samsung_ms_prod['Week33']
```

```python
#내림차순으로 정렬
samsung_ms_line = samsung_ms_prod.sort_values(['Weight_sum'], ascending=[False])
samsung_ms_line
```

```python
#Weight_sum탈락시키기
samsung_ms_line_review_count = samsung_ms_line.drop(samsung_ms_line.columns[-1:], axis='columns')
samsung_ms_line_review_count
```

```python
data = pd.DataFrame(samsung_ms_line)
data.to_csv('flipkart_line_weight_sum#.csv')
```

```python
#파이차트 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rc

%matplotlib inline
```

```python
#weigh_sum만 불러오기
samsung_ms_line_ws = samsung_ms_line[['Weight_sum']]
samsung_ms_line_ws
```

```python
#파이차트 그리기
plt.pie(samsung_ms_line_ws,
        explode = (0.05, 0, 0, 0, 0),
        labels = ['ON series', 'A series', 'J series', 'M series', 'S series', ],
        colors = ['orange', 'dodgerblue', '#F44E54', 'green', '#76AD3B'], 
        startangle = 180,
        autopct = '%1.2f%%')

plt.rcParams.update({'font.size': 25})
plt.rcParams['figure.figsize'] = [20, 15]
plt.axis('equal', fontsize=18)
plt.title('Flipkart Samsung Smartphone Review Share', fontsize = 40)       
plt.legend(fontsize=22, loc='upper left')

plt.show()
```

```python
#백분율 구하기 vendor_weigh_2
samsung_ms_line_per = (samsung_ms_line[0:] / samsung_ms_line[0:].sum()) * 100
samsung_ms_line_per
```

```python
# csv파일로 저장하기 (주별 리뷰 점유율 percentage)
data1 = pd.DataFrame(samsung_ms_line_per)
data1.to_csv('weekly_flipkart_line_weight_sum_percentage#.csv')
```

```python
#행과 열을 바꾸기
samsung_ms_line_per = np.transpose(samsung_ms_line_per)
samsung_ms_line_per.head()
```

```python
# 누적막대그래프 만들기
samsung_ms_line_per.plot.bar(stacked=True, fontsize=25,
                          colors = ['orange', 'dodgerblue', '#F44E54', 'green', '#76AD3B'], alpha = 0.7)

# Set the title and labels
plt.rcParams['figure.figsize'] = [15, 13]
plt.legend(fontsize=22, loc = 'lower right')
plt.xlabel('weeks', fontsize=23)
plt.ylabel('Percentage', fontsize=23)
plt.title('Flipkart Weekly Samsung Smartphone Review Share ', fontsize=40)


# show the plot
plt.show()
```

```python
x= samsung_ms_line_review_count.columns[0:]
y=  samsung_ms_line_review_count[0:]

pal = ['orange', 'dodgerblue', '#F44E54', 'green', '#76AD3B']
plt.stackplot(x,y, labels=['ON series', 'A series', 'J series', 'M series', 'S series', ], 
              colors=pal, alpha=0.7 )

plt.title("Flipkart Weekly Samsung Smartphone Review Count", fontsize=40, fontweight=0, color='black')
plt.xlabel("Weeks", fontsize = 24)
plt.xticks(fontsize = 24)

plt.ylabel("Review counts", fontsize = 24)
plt.yticks(fontsize = 24)

plt.legend(loc='upper left', fontsize = 22)
plt.show()
 
```
